```python
from argparse import ArgumentParser
import os
from speechkit import model_repository, configure_credentials, creds
from dotenv import load_dotenv
from scipy.io.wavfile import read

load_dotenv()

# Authentication via an API key.
configure_credentials(
    yandex_credentials=creds.YandexCredentials(
        iam_token=os.getenv("YANDEX_IAM_TOKEN")
    )
)


headers = {
    "x-folder-id": os.getenv("YANDEX_FOLDER_ID"),
}

def synthesize(text, export_path):
    model = model_repository.synthesis_model(custom_headers=headers)

    # Set the synthesis settings.
    model.voice = "nigora"

    # Performing speech synthesis and creating an audio file with results.
    result = model.synthesize(text, raw_format=False)
    result.export(export_path, "wav")

    return read(export_path)[1]
```

```python
from IPython.display import Audio
```

```python
Audio(synthesize(text="Assalom Aleykum", export_path="output.wav")[1], rate=44100)
```
